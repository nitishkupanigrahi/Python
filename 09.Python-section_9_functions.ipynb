```python
def add(a, b):
    return a + b
def sub(a, b):
    return a - b
def mul(a, b):
    return a * b
def div(a, b):
    if b == 0:
        return "Error: Division by zero is not allowed."
    return a / b
def calculator():
    print("Welcome to the calculator!")
    a = float(input("Enter the first number: "))
    b = float(input("Enter the second number: "))
    print("Options:")
    print("1: Add")
    print("2: Subtract")
    print("3: Multiply")
    print("4: Divide")
    option = int(input("Choose an option (1-4): "))
    if option == 1:
        print(f"The result is: {add(a, b)}")
    elif option == 2:
        print(f"The result is: {sub(a, b)}")
    elif option == 3:
        print(f"The result is: {mul(a, b)}")
    elif option == 4:
        print(f"The result is: {div(a, b)}")
    else:
        print("Invalid option. Please choose a valid option (1-4).")
calculator()
```
